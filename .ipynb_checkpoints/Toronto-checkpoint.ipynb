<h1>Toronto Neighborhood Project</h1>


Import necessary libraries/modules

```python
import requests
import bs4
import pandas as pd
import numpy as np
```

Get and process data from the web

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = requests.get(url).text
page_bs = bs4.BeautifulSoup(page)
table = page_bs.select("table")
rows = table[0].select("tr")[1:]
```

Process data with beautiful soup and create dataframe object

```python
data = []
for row in rows:
    td = row.select('td')
    if td[1].text != "Not assigned\n":
        postCode = td[0].text.strip()
        borough = td[1].text.strip()
        neighborhood = td[2].text.strip()
        
        if neighborhood == "Not assigned":
            neighborhood = borough
            
        new_row = [postCode, borough, neighborhood]
        data.append(new_row)
df = pd.DataFrame(data)
df.columns=['PostalCode','Borough','Neighborhood']
```

Get latitude / longitude data

```python
geo_data = pd.read_csv("Geospatial_Coordinates.csv")

latitudes = []
longitudes = []

for row in range(len(df)):
    latitudes.append(geo_data[geo_data['Postal Code'] == df.iloc[row].PostalCode].Latitude.value)
    longitudes.append(geo_data[geo_data['Postal Code'] == df.iloc[row].PostalCode].Longitude.value)

df['Latitudes'] = latitudes
df['Longitudes'] = longitudes
```
